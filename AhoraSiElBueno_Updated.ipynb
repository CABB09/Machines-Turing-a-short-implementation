```python
#Diccionario de las codificaciones de los movimientos

movimientos = { "0": "0", "10": "1", "R" :"110", "L" : "1110", "STOP": "11110"}
```

```python
def decimal_to_binary_simplified(decimal_number):
    """ Convierte un número decimal en una cadena binaria simplificada. """
    return bin(int(decimal_number))[2:]  # Se usa [2:] para eliminar el prefijo '0b' de la cadena binaria.

def binary_to_decimal(binary):
    return int(binary, 2)

def display_machine_info(machine_id):
    if machine_id in turing_machines:
        machine = turing_machines[machine_id]
        decimal_value = binary_to_decimal(machine["binary"])
        decimal_value_simpli = binary_to_decimal(machine["binary_simplified"])
        print(f"Machine {machine_id}:")
        print(f"  Instructions: {machine['instructions']}")
        print(f"  Binary: {machine['binary']}")
        print(f" Binary Simplified : {machine['binary_simplified']}")
        print(f"  Decimal Binary: {decimal_value}")
        print(f" Decimal Binary Simplified: {decimal_value_simpli}")
    else:
        print("Machine not found.")
```

```python
import re #A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.

def find_correct_machines():
    correct_machines = {}
    for machine_id, machine_info in turing_machines.items():
        if "STOP" in machine_info["instructions"]:
            correct_machines[machine_id] = machine_info
    return correct_machines

def simplified_to_full_binary(simplified_binary):
    """ Añade 'R' al inicio y al final de la codificación simplificada y convierte a binario completo. """
    return movimientos["R"] + simplified_binary + movimientos["R"]

def fill_movements(binary_code):
    """ Rellena la codificación binaria con los símbolos de movimientos utilizando expresiones regulares. """
    # Reemplazar STOP antes de L y R porque es la secuencia más larga y evitaría conflictos de superposición.
    filled_binary = re.sub(r"11110", "STOP", binary_code)
    filled_binary = re.sub(r"1110", "L", filled_binary)
    filled_binary = re.sub(r"110", "R", filled_binary)
    filled_binary = re.sub(r"10", "1", filled_binary)
    filled_binary = re.sub(r"0", "0", filled_binary)
    return filled_binary
```

```python
def generate_rules(filled_binary):
    """Genera un diccionario de transiciones a partir de la cadena binaria llena de movimientos."""
    rules_with_commas = re.sub(r"(R|L|STOP)", r"\1,", filled_binary)
    rules_with_commas = re.sub(r"(R|L|STOP),(?=(R|L|STOP))", r"\1,", rules_with_commas)
    rules_cleaned = re.sub(r",+", ",", rules_with_commas).strip(",")
    rules_list = rules_cleaned.split(",")

    program = {}
    state = 0

    for rule in rules_list:
        if rule in ['R', 'L', 'STOP']:
            current_symbol = 0
            new_symbol = 0
            movement = rule
        else:
            match = re.match(r"(\d*)(\d)(R|L|STOP)", rule)
            if match:
                nums, last_digit, movement = match.groups()
                current_symbol = int(last_digit)
                new_symbol = int(last_digit) if nums == "" else int(nums)
            else:
                continue

        if movement == 'STOP':
            movement = 'H'
        new_state = state + 1 if movement != 'H' else state

        if (state, current_symbol) in program:
            state = max(state, max([s for s, _ in program.keys()])) + 1

        program[(state, current_symbol)] = (new_state, new_symbol, movement)
        
        if movement != 'H':
            state = new_state

    return program
```

```python
def single_instruction_format(state_from_read, state_to_write_direction):
    state_from, read = state_from_read
    state_to, write, direction = state_to_write_direction
    direction_map = {-1: 'L', 0: 'H', 1: 'R', 'L': 'L', 'H': 'H', 'R': 'R'}
    return {
        "StateFrom": state_from,
        "Read": read,
        "StateTo": state_to,
        "Write": write,
        "Direction": direction_map[direction]
    }

def preprocess_program(program):
    sorted_program = sorted(program.items(), key=lambda x: x[0])
    return sorted_program

def turing_program_format(program):
    preprocessed_program = preprocess_program(program)
    return [single_instruction_format(key, value) for key, value in preprocessed_program]

def turing_machine(state, tape, transitions):
    head_position = state['pos']
    while True:
        current_symbol = tape[head_position]
        action = transitions.get((state['state'], current_symbol))
        if action is None:
            break
        new_state, new_symbol, move = action
        tape[head_position] = new_symbol
        state['state'] = new_state
        if move == 'R':
            head_position += 1
        elif move == 'L':
            head_position -= 1
        elif move == 'H':
            break
        if head_position < 0 or head_position >= len(tape):
            tape.append(0)
    return tape
```

```python
def log_turing_machine(state, tape, transitions):
    """
    Simula una máquina de Turing y registra cada paso.

    Parameters:
    state (dict): Estado inicial de la máquina.
    tape (list): Cinta inicial de la máquina.
    transitions (dict): Transiciones de la máquina.

    Returns:
    list: Cinta resultante después de ejecutar la máquina de Turing.
    """
    head_position = state['pos']
    steps = []
    while True:
        current_symbol = tape[head_position]
        action = transitions.get((state['state'], current_symbol))
        if action is None:
            break
        new_state, new_symbol, move = action
        tape[head_position] = new_symbol
        steps.append((state['state'], current_symbol, new_state, new_symbol, move, head_position))
        state['state'] = new_state
        if move == 'R':
            head_position += 1
        elif move == 'L':
            head_position -= 1
        elif move == 'H':
            break
        if head_position < 0:
            tape.insert(0, 0)
            head_position = 0
        elif head_position >= len(tape):
            tape.append(0)
    return tape, steps

def print_steps(steps):
    """
    Imprime los pasos de la simulación de la máquina de Turing.

    Parameters:
    steps (list): Lista de pasos registrados durante la simulación.
    """
    for i, step in enumerate(steps):
        state_from, read, state_to, write, direction, pos = step
        print(f"Paso {i+1}: Estado {state_from}, Cabeza en {pos}, Leyendo {read} -> Estado {state_to}, Escribiendo {write}, Moviendo {direction}")
```
